```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# to Classify BMI
def classify_bmi(bmi_value):
    if bmi_value < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi_value < 25:
        return 'Normal weight'
    elif 25 <= bmi_value < 30:
        return 'Overweight'
    else:
        return 'Obese'

bmi_categories = [classify_bmi(x) for x in bmi]
bmi_df = pd.DataFrame({'BMI': bmi, 'Category': bmi_categories})

# Plot
plt.figure(figsize=(10, 6))
sns.countplot(data=bmi_df, x='Category', order=['Underweight', 'Normal weight', 'Overweight', 'Obese'])
plt.title('BMI Classification Distribution')
plt.xlabel('BMI Category')
plt.ylabel('Count')
plt.show()
```
